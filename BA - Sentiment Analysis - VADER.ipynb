```python
# It is part of the NLTK library, and its primary focus is to assign sentiment scores to text based on both the words and context.
# The compound score ranges from -1 (most negative) to +1 (most positive), with 0 being neutral.
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize analyzer
sia = SentimentIntensityAnalyzer()

# Sample text
text = "The movie was amazing, but the ending was a bit disappointing."

# Perform sentiment analysis
scores = sia.polarity_scores(text)
print("Sentiment Scores:", scores)  # Contains 'pos', 'neu', 'neg', and 'compound -which is the average of the 3'
```

```python
# run the polarity score on the entrire dataset
result = {}
for i, row in tqdm(df.iterrows(), total = len(df)):
    text = row ['Comment']
    rate = row ['Rating']
# Fetch the actual text from the 'Comment' column
    result [text] = sia.polarity_scores(text)  # Pass the actual text to SIA
```

```python
result
```

```python
# Convert results into a DataFrame. the T is to transpose the results 
sentiment_df = pd.DataFrame(result).T.reset_index().rename(columns = {'index' : 'Comment'})
merged_df = sentiment_df.merge(df, how ='left')
print(merged_df.tail())
```

```python
#plot merged values
ax = sns.boxplot(data = merged_df, x= 'Rating', y= 'compound')
ax.set_title('Compound score by British Airline reviews')
plt.show()
```

```python
fig, axs = plt.subplots (1,3, figsize = (15,3))
sns.boxplot(data = merged_df, x= 'Rating', y= 'pos', ax= axs[0])
sns.boxplot(data = merged_df, x= 'Rating', y= 'neu', ax= axs[1])
sns.boxplot(data = merged_df, x= 'Rating', y= 'neg', ax= axs[2])
axs[0].set_title ('Positive')
axs[1].set_title ('Neutral')
axs[2].set_title ('Negative')
plt.show()
#the vader model seems to be working well. For customers gave a high rating, the sentiment score is higher than the those who rated lower.
# Which is what is expected
```

Conclusion; 
VADER is a lightweight, fast, and effective tool for sentiment analysis on informal and social media text, offering robust features like emoji handling and punctuation-based sentiment adjustment. However, it has limitations in its ability to understand context, sarcasm, and domain-specific language. For simple applications, it's an excellent choice, but for more nuanced analysis, especially on formal text or complex sentiment, more advanced methods may be required.
